```julia
struct SimpleOG
    B
    M
    α
    β
    R
    Q
end

function SimpleOG(;B=10, M=5, α=0.5, β=0.9)

    u(c) = c^α
    n = B + M + 1
    m = M + 1

    R = Matrix{Float64}(n, m)
    Q = zeros(Float64,n,m,n)

    for a in 0:M
        Q[:, a + 1, (a:(a + B)) + 1] = 1 / (B + 1)
        for s in 0:(B + M)
            R[s + 1, a + 1] = a<=s ? u(s - a) : -Inf
        end
    end

    return SimpleOG(B, M, α, β, R, Q)
end
```

```julia
using QuantEcon
g = SimpleOG()
ddp = DiscreteDP(g.R, g.Q, g.β)
results = solve(ddp, PFI)
fieldnames(results)
```
